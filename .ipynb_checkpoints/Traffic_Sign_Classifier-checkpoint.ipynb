````python
## Todo
# Augment the dataset by rotating or shifting images or by changing colors



import sys
import numpy as np
import pickle
import matplotlib.pyplot as plt
import os.path

'''
Code in image_effect function modified from image effect functions obtained from the below URL
https://github.com/eqbal/CarND-Traffic-Sign-Classifier-Project/blob/master/image_effect.py

'''

def image_effect(img, image_effect=None):
        
        #select a random effect if none specified
        if image_effect == None:
            image_effect = random.choice(['Translate','Rotate','Shear','g_Blur','Gamma'])
            
        
        
        #Brightness
        if image_effect == 'Brightness':
            img = cv2.cvtColor(img,cv2.COLOR_RGB2HSV)
            random_bright = .25+np.random.uniform()
            img[:,:,2] = img[:,:,2]*random_bright
            img = cv2.cvtColor(img,cv2.COLOR_HSV2RGB)
            augmentation_stats['Brightness'] +=1
        
        #Translate
        if image_effect == 'Translate':
            x = img.shape[0]
            y = img.shape[1]

            x_trans = np.random.uniform(-0.2 * x, 0.2 * x)
            y_trans = np.random.uniform(-0.2 * y, 0.2 * y)

            trans_matrix = np.float32([[1, 0, x_trans], [0, 1, y_trans]])
            trans_img = cv2.warpAffine(img, trans_matrix, (x, y))
            img = trans_img
            augmentation_stats['Translate'] +=1
        
        #Rotate
        if image_effect == 'Rotate':
            row, col, channel = img.shape
            angle = np.random.uniform(-45, 45)
            rotation_point = (row / 2, col / 2)
            rotation_matrix = cv2.getRotationMatrix2D(rotation_point, angle, 1)
            rotated_img = cv2.warpAffine(img, rotation_matrix, (col, row))
            img = rotated_img
            augmentation_stats['Rotate'] +=1
        
        #Shear
        if image_effect == 'Shear':
            x, y, channel = img.shape
            shear = np.random.randint(5,15)
            pts1 = np.array([[5, 5], [20, 5], [5, 20]]).astype('float32')
            pt1 = 5 + shear * np.random.uniform() - shear / 2
            pt2 = 20 + shear * np.random.uniform() - shear / 2
            pts2 = np.float32([[pt1, 5], [pt2, pt1], [5, pt2]])
            M = cv2.getAffineTransform(pts1, pts2)
            result = cv2.warpAffine(img, M, (y, x))
            img = result
            augmentation_stats['Shear'] +=1
            
        #Gaussian_Blur
        if image_effect == 'g_Blur':
            r_int = np.random.randint(0, 2)
            odd_size = 2 * r_int + 1
            img = cv2.GaussianBlur(img, (odd_size, odd_size), 0)
            augmentation_stats['Blur'] +=1
            
        #Median Blur
        if image_effect == 'm_Blur':
            r_int = np.random.randint(0, 2)
            odd_size = 2 * r_int + 1
            img = cv2.medianBlur(img, (odd_size, odd_size), 0)
            augmentation_stats['Blur'] +=1
        
        
        #Gamma
        if image_effect == 'Gamma':
            gamma = np.random.uniform(0.3, 1.5)
            invGamma = 1.0 / gamma
            table = np.array([((i / 255.0) ** invGamma) * 255 for i in np.arange(0, 256)]).astype("uint8")
            img = cv2.LUT(img, table)
            augmentation_stats['Gamma'] +=1
        

        
        return img
````
